Cleaning up dataframes: dropping unnessecary columns

```python
## Zillow data: goal is to track the avg house value in each zipcode by month over three year period
tx_housing_data = housing_data[(housing_data['State'] == 'TX') & (housing_data['City'] == 'Austin')]

zipcodes = tx_housing_data['RegionName'].unique()
tx_housing_data.head()
drop_years = ['2000', '2001', '2002', '2003', '2004', '2005', '2006', 
'2007', '2008', '2009', '2010', '2011', '2012', '2013', '2014', '2015', '2016', '2017', '2018', '2019']


for i in range(len(drop_years)):
    tx_housing_data = tx_housing_data.drop(tx_housing_data.filter(regex=drop_years[i]).columns, axis=1)
tx_housing_data.head()
# drop unneccessary columns

drop_cols = ['RegionID', 'SizeRank', 'RegionType', 'StateName', 'State', 'City']
aus_housing = tx_housing_data.drop(drop_cols, axis=1)
aus_housing.head()
aus_housing_final = aus_housing
```

```python
# Number of Evictions per month per zipcode
eviction_data.head()
eviction_temp = eviction_data[['GEOID', 'month', 'filings_2020']]
eviction_temp.head()
final_eviction = eviction_temp
```

```python
#Number of homeless in shelters over 5 years in Austin (2016 - 2021)
homeless_data.head()
homeless_data['Demographic Category'].unique()
new_df = homeless_data[homeless_data['Demographic Category'] == 'Race']
homeless_drop = ['Demographic Category', 'Fiscal Year Start', 'Fiscal Year End']
homeless_final = new_df.drop(homeless_drop, axis=1)
homeless_final.head()
```

```python
## Census Data

med_2018 = pd.read_csv('2018_med_inc.csv')
med_2019 = pd.read_csv('2019_med_inc.csv')
med_2020 = pd.read_csv('2020_med_inc.csv')
med_2021 = pd.read_csv('2021_med_inc.csv')
#census_df = census_data.dropna(axis=1)
#census_df.head()
```

```python
# getting rid of uncessesary columns
df_list = [med_2018, med_2019, med_2020, med_2021]
med_2018 = med_2018.iloc[:, [1, 2]]
med_2019 = med_2019.iloc[:, [1, 2]]
med_2020 = med_2020.iloc[:, [1, 2]]
med_2021 = med_2021.iloc[:, [1, 2]]
```

```python
# getting rid of ZCTA5
med_2018['NAME'] = med_2018['NAME'].str.replace('ZCTA5', '')
med_2019['NAME'] = med_2019['NAME'].str.replace('ZCTA5', '')
med_2020['NAME'] = med_2020['NAME'].str.replace('ZCTA5', '')
med_2021['NAME'] = med_2021['NAME'].str.replace('ZCTA5', '')
```

```python
#getting rid of first row
final_med_2018 = med_2018.tail(-1)
final_med_2019 = med_2019.tail(-1)
final_med_2020 = med_2020.tail(-1)
final_med_2021 = med_2021.tail(-1)
```

```python
#Final Median Income CSV
final_med_2018.to_csv('../Final_Resources/final_med_2018.csv')
final_med_2019.to_csv('../Final_Resources/final_med_2019.csv')
final_med_2020.to_csv('../Final_Resources/final_med_2020.csv')
final_med_2021.to_csv('../Final_Resources/final_med_2021.csv')
```
